```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Average metrics data
data = {
    "Scenario": ["Scenario 1 (priority_score)", "Scenario 1 (problem_domain)", "Scenario 2 (priority_score)", "Scenario 2 (problem_domain)"],
    "Precision": [0.41, 0.48, 0.41, 0.87],
    "Recall": [0.38, 0.49, 0.37, 0.60],
    "F1-score": [0.35, 0.47, 0.36, 0.63],
    "Accuracy": [0.41, 0.77, 0.42, 0.85]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Melt the DataFrame for easier visualization
df_melted = pd.melt(df, id_vars="Scenario", var_name="Metric", value_name="Value")

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x="Scenario", y="Value", hue="Metric", data=df_melted, palette="Set2")

# Annotate bars with exact values
for index, row in df.iterrows():
    for i, metric in enumerate(["Precision", "Recall", "F1-score", "Accuracy"]):
        plt.text(index + i * 0.2 - 0.3, row[metric] + 0.02, f"{row[metric]:.2f}", ha='center')

plt.title('Average Metrics Comparison Across Scenarios')
plt.ylabel('Score')
plt.xticks(rotation=45, ha="right")
plt.ylim(0, 1)  # Set y-axis limit to ensure annotations are visible
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data provided
data = {
    "Metric": ["Precision", "Recall", "F1-score", "Accuracy"],
    "Scenario 2": [0.41, 0.37, 0.36, 0.42],
    "Other Research (Andrian et al., 2022)": [0.74, 0.72, 0.72, 0.74],
    "Other Research (Suprayogi et al., 2022)": [1.00, 0.78, 0.90, 0.85]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the positions and width for the bars
positions = range(len(df))
width = 0.2

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

for i, (metric, color) in enumerate(zip(df.columns[1:], ['b', 'g', 'r'])):
    bars = ax.bar([p + i * width for p in positions], df[metric], width=width, label=metric, color=color)

    # Add text labels above the bars
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black', fontsize=8)

ax.set_xticks([p + width for p in positions])
ax.set_xticklabels(df['Metric'])
ax.set_ylabel('Scores')
ax.set_title('Comparison of Metrics Across Scenarios and Studies')
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```
